```python
import pandas as pd
```

```python
df = pd.read_csv('countries of the world.csv')
pd.set_option('display.max_rows', None)
```

```python
df.info()
```

```python
df['Area (sq. mi.)'].value_counts()
```

```python
def area_km(value):
    return value*1.60934 
df['Area (sq. mi.)'] = df['Area (sq. mi.)'].apply(area_km)
```

```python
df['Area (sq. mi.)'].value_counts()
```

```python
df['Pop. Density (per sq. mi.)'].value_counts()
```

```python
def dens_conv(value):
    result = value.replace(',', '.')
    return float(result) * 1.60934 
df['Pop. Density (per sq. mi.)'] = df['Pop. Density (per sq. mi.)'].apply(dens_conv)
```

```python
df['Pop. Density (per sq. mi.)'].value_counts()
```

```python
df.info()
```

```python
df['Coastline (coast/area ratio)'].value_counts()
```

```python
df['Coastline (coast/area ratio)'] = df['Coastline (coast/area ratio)'].apply(dens_conv)
```

```python
df['Coastline (coast/area ratio)'].value_counts()
```

```python
df['Net migration'].value_counts()
```

```python
def net_m(value):
    if isinstance(value, str):
        value = value.replace(',', '.')
        value = float(value)*1000 
    else:
        value = float(value)*1000
    return value 
df['Net migration'] = df['Net migration'].apply(net_m)
```

```python
df['Net migration'].value_counts()
```

```python
df.info()
```

```python
mean = df['GDP ($ per capita)'].mean()
df['GDP ($ per capita)'] = df['GDP ($ per capita)'].fillna(mean)
```

```python
df.info()
```

```python
mean = df['GDP ($ per capita)'].mean()
print(mean)
```

```python
df_rich = df[df['GDP ($ per capita)'] >= mean]
df_poor = df[df['GDP ($ per capita)'] < mean]
df_rich.count()
df_poor.count()
```

```python
df['Literacy (%)'].value_counts()
```

```python
def literacy(value):
    if isinstance(value, str):
        value = value.replace(',', '.')
    return float(value)
df['Literacy (%)'] = df['Literacy (%)'].apply(literacy)
```

```python
df['Literacy (%)'].value_counts()
```

```python
df.info()
```

```python
mean = df['Literacy (%)'].mean()
df['Literacy (%)'] = df['Literacy (%)'].fillna(mean)
```

```python
df.info()
```

```python
mean = df['GDP ($ per capita)'].mean()
df_rich = df[df['GDP ($ per capita)'] >= mean]
df_poor = df[df['GDP ($ per capita)'] < mean]
rich_lit = df_rich['Literacy (%)'].mean()
poor_lit = df_poor['Literacy (%)'].mean()
print(rich_lit)
print(poor_lit)
```

```python
df_rich.plot(kind = 'barh', x = 'Country', y = 'Literacy (%)', figsize = (10, 12), grid = True)
```

```python
df_poor.plot(kind = 'barh', x = 'Country', y = 'Literacy (%)', figsize = (10, 12))
```
